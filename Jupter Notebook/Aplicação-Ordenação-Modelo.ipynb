```python
#Importa os objetos com o treinamento do Random Forest
Modelo = pd.read_pickle(r'D:\Mestrado\MasterMedicalDecisionSupport\Arquivos_Tot\Model_MDS_4-03.pickle')

def processing_patients(*patients):
    # Verificar quantidade de pacientes passados
    if len(patients) < 2:
        print(patients)
        return tuple([patients[i]] for i in range(0, len(patients)))
    
    # Fazer separação dos pacientes em pares de forma que todos os pacientes sejam comparados com os outros
    pairs = [(patients[i], patients[j]) for i in range(len(patients)) for j in range(i+1, len(patients))]
    
    # Concatenar a cada 2 pacientes os dados
    results = []
    for patient1, patient2 in pairs:
        print('Pacientes:',patient1[:1], ' e ', patient2[:1]) # Trata o nome:
        data_prepared = [y for x in [patient1[1:], patient2[1:]] for y in x]
        
        # Fazer a previsão
        comparation = Modelo.predict([data_prepared])
        print('resultado: ', comparation)
        
        # se a comparação for der a resposta 1, é o paciente 1 que é mais prioritário
        # se a comparação for der a resposta 2, é o paciente 2 que é mais prioritário
        # O paciente não pode repetir na listagem final
        if comparation == 1:
            results.append((patient1, 1))
        else:
            results.append((patient2, 1))
    
    # Ordenar a resposta conforme a prioridade do paciente
    unique_results = []
    for r in results:
        if r[0] not in unique_results:
            unique_results.append(r[0])
    ranked_results = []
    for p in patients:
        total_points = 0
        for r in results:
            if r[0] == p:
                total_points += r[1]
        ranked_results.append((p, total_points))
    ranked_results.sort(key=lambda x: x[1], reverse=True)
    
    return tuple([r[0] for r in ranked_results])
```
